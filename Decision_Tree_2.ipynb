```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/BDA/Parkinsons_big data.csv')
df
```

```python
df.isnull().sum()
```

```python
df['MDVP:Flo(Hz)']=df['MDVP:Flo(Hz)'].fillna(df['MDVP:Flo(Hz)'].mean())
```

```python
df['MDVP:Jitter(Abs)']=df['MDVP:Jitter(Abs)'].fillna(df['MDVP:Jitter(Abs)'].mean())
```

```python
df['MDVP:APQ']=df['MDVP:APQ'].fillna(df['MDVP:APQ'].mean())
```

```python
df['D2']=df['D2'].fillna(df['D2'].mean())
```

```python
df.isnull().sum()
```

```python
features=df.loc[:,df.columns!='status'].values[:,1:]
labels=df.loc[:,'status'].values
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler((-1,1))
X = scaler.fit_transform(features)
y = labels
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state = 7)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion='entropy',max_depth=10,random_state=7)
dt.fit(X_train,y_train)
```

```python
y_pred = dt.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_pred,y_test)*100)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/ML/Boston.csv')
df
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.corr()
```

```python
X = df.drop(['medv'],axis=1).values
y = df['medv'].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state = 7)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_train=ss.fit_transform(X_train)
X_test=ss.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor(criterion = 'squared_error',max_depth=10,random_state=7)
dtr.fit(X_train,y_train)
```

```python
y_pred = dtr.predict(X_test)
```

```python
from sklearn.metrics import r2_score as score
print(score(y_pred,y_test)*100)
```

```python
from sklearn.metrics import mean_squared_error as mse
mse(y_pred,y_test)
```
